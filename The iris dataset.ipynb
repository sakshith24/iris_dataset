# MODELING - SIMPLE MODEL
BY LOOKING AT THE DATA 

```python
def single_feature_prediction(petal_length):
    if petal_length < 2.5:
        return 0
    elif petal_length < 4.7:
        return 1
    else:
        return 2
```

```python
df_train.head()
```

```python
x_train[:,2]
```

```python
manual_y_prediction = np.array([single_feature_prediction(val) for val in x_train[:,2]])
```

```python
manual_mode_accuracy = np.mean(manual_y_prediction == y_train)
```

```python
print(f"manual accuracy : {manual_mode_accuracy * 100}")
```

# modelling through logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

# using validation set to evaluate the model

```python
# xt stands for x_train, xv for x_validation
xt,xv,yt,yv = train_test_split(x_train,y_train, test_size=0.25) 
```

```python
model = LogisticRegression()
```

```python
model.fit(xt,yt)
```

```python
y_pred = model.predict(xv)
```

```python
np.mean(y_pred == yv)
```

```python
model.score(x_train,y_train) # we will never use this because it is memorized
```

# using cross validation to evaluate the model

```python
from sklearn.model_selection import cross_val_predict,cross_val_score
```

```python
model = LogisticRegression(max_iter =200)
```

```python
accuracies = cross_val_score(model,x_train,y_train,cv=5 , scoring= "accuracy")
```

```python
np.mean(accuracies)
```

```python
y_pred = cross_val_predict(model,x_train,y_train,cv=5)
```

```python
masked_pred_correctly = y_pred == y_train
```

```python
not_masked_correctly = ~masked_pred_correctly
```

```python
x_train[not_masked_correctly]
```

```python
df_train.head()
```

```python
df_prediction = df_train.copy()
```

```python
df_prediction["correct prediction"] = masked_pred_correctly
```

```python
df_prediction.head()
```

```python
df_prediction["prediction"] = y_pred
```

```python
df_prediction["prediction_label"] = df_prediction["prediction"].map({0:'setosa',1:'versicolor',2:'virginica'})
```

```python
sns.scatterplot(x="petal length (cm)" , y = "petal width (cm)",hue = "prediction_label", data = df_prediction)
```

```python
sns.scatterplot(x="petal length (cm)" , y = "petal width (cm)",hue = "target_name", data = df_prediction)
```

```python
def plot_incorrect_predictions(df_prediction,x_axis,y_axis):
    fig,axs = plt.subplots(2,2, figsize =(10,10))
    axs = axs.flatten()
    sns.scatterplot(x="petal length (cm)" , y = "petal width (cm)",hue = "prediction_label", data = df_prediction ,ax=axs[0] )
    sns.scatterplot(x="petal length (cm)" , y = "petal width (cm)",hue = "target_name", data = df_prediction,ax=axs[1])
    sns.scatterplot(x="petal length (cm)" , y = "petal width (cm)",hue = "correct prediction", data = df_prediction,ax=axs[2])

    plt.show()
    
```

```python
plot_incorrect_predictions(df_prediction,"petal length (cm)" , "petal width (cm)")
```
